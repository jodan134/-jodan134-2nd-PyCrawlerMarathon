```python
import requests
import json
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json
count = 0
r = requests.get('https://www.dcard.tw/_api/forums/2019_ncov/posts?popular=true')
data = json.loads(r.text)
for d in data:
    count = count + 1

print("一共"+ str(count) + "筆資料")

for d in data:
    for tag in d:
        print(tag)
    break
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
cnt = 0
for d in data:
    cnt = cnt + 1
    print("第" + str(cnt) + "筆資料:")  
    print ("標題:" + d['title'], "貼文時間:" + d['createdAt'],"留言人數:"+ str(d['likeCount']), "按讚人數:"+ str(d['commentCount']),sep="|")
    print()
    
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

popular = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=true')
unpopular = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=false')

dataPop = json.loads(popular.text)
dataUn = json.loads(unpopular.text)

likeCnt = 0
commentCnt = 0
postCnt = 0

for d in dataPop:
    likeCnt = likeCnt + d['likeCount']
    commentCnt = commentCnt + d['commentCount']
    postCnt = postCnt + 1

print("武漢肺炎熱門文章")
print("平均按讚人數:", likeCnt/postCnt)
print("平均留言人數:",commentCnt/postCnt)
print('\n')

likeCnt = 0
commentCnt = 0
postCnt = 0
for d in dataUn:
    likeCnt = likeCnt + d['likeCount']
    commentCnt = commentCnt + d['commentCount']
    postCnt = postCnt + 1
    
print("武漢肺炎非熱門文章")
print("平均按讚人數:", likeCnt/postCnt)
print("平均留言人數:",commentCnt/postCnt)
print('\n')
```
